```python
from autogluon.tabular import TabularDataset, TabularPredictor
data = TabularDataset(df)
data
```

```python
data['Rupture_Results'] = data['Rupture_Results'].str.strip()
```

```python
# Split the data into train and test datasets
from sklearn.model_selection import train_test_split

D_train, D_test = train_test_split(data, test_size=0.20, random_state=42)
```

```python
predictor = TabularPredictor(
    label="Rupture_Results",
    problem_type="binary",
).fit(D_train, time_limit=60, presets="best_quality")
```

```python
test_data = TabularDataset(D_test)

```

```python
test_data
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# Assuming you have a DataFrame 'test_data' and a model 'predictor'
label = 'Rupture_Results'  # Column name for actual outcomes

# Predicting outcomes
y_pred = predictor.predict(test_data.drop(columns=[label]))

# Assuming the actual labels are in test_data under the column 'Rupture_Results'
y_true = test_data[label]

# Define the positive class label
positive_class_label = 'No Ruptures & Rebounds'

# Calculate evaluation metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, pos_label=positive_class_label)
recall = recall_score(y_true, y_pred, pos_label=positive_class_label)
f1 = f1_score(y_true, y_pred, pos_label=positive_class_label)
conf_matrix = confusion_matrix(y_true, y_pred)


print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:\n", conf_matrix)


```
